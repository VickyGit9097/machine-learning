```python
import pandas as pd
```

```python
data=pd.read_csv("salaries.csv")
data
```

```python
data['company']=data['company'].astype('category')
data['company']=data['company'].cat.codes

data['job']=data['job'].astype('category')
data['job']=data['job'].cat.codes

data['degree']=data['degree'].astype('category')
data['degree']=data['degree'].cat.codes
```

```python
data
```

```python
x=data.drop(columns='salary_more_then_100k')
x
```

```python
y=data['salary_more_then_100k']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
y_train_pred=model.predict(x_train)
y_train[:5]
```
